```python
# Sample dataset 1 - Sales Data

import pandas as pd

data_sales = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"],
    "Product": ["A", "B", "A", "C", "B"],
    "Quantity": [10, 5, 8, 6, 7],
    "Price": [100, 200, 100, 300, 200],
    "Customer_ID": [101, 102, 103, 104, 101],
}
df_sales = pd.DataFrame(data_sales)
print(df_sales)
```

```python
pip install matplotlib
```

```python
"""Task 1: Plot a bar chart showing total sales (Quantity * Price) by Product."""

import pandas as pd
import matplotlib.pyplot as plt

df_sales['Total_Sales'] = df_sales['Quantity'] * df_sales['Price']

sales_by_product = df_sales.groupby('Product')['Total_Sales'].sum()

plt.figure(figsize=(8, 5))
sales_by_product.plot(kind='bar', color='green')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
"""Task 2: Create a line plot showing the trend of Temperature over time."""

# Sample dataset 3 - Weather Data
data_weather = {
    "Date": ["2023-09-01", "2023-09-02", "2023-09-03", "2023-09-04", "2023-09-05"],
    "City": ["New York", "Los Angeles", "New York", "Chicago", "Los Angeles"],
    "Temperature": [80, 85, 82, 75, 88],
    "Rainfall": [0.1, 0.0, 0.2, 0.0, 0.0],
}
df_weather = pd.DataFrame(data_weather)
print(df_weather)

```

```python
plt.figure(figsize=(8, 5))
plt.plot(df_weather['Date'], df_weather['Temperature'], marker='o', linestyle='-', color='blue', label='Temperature')

plt.title('Temperature over time')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```

```python
"""Task 3: Plot a scatter plot between two numerical columns (e.g., Salary vs Employee_ID)."""

# Sample dataset 2 - Employee Data
data_employee = {
    "Employee_ID": [1, 2, 3, 4, 5],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department": ["HR", "IT", "HR", "Sales", "IT"],
    "Salary": [50000, 60000, 55000, 70000, 62000],
}
df_employee = pd.DataFrame(data_employee)
print(df_employee)

```

```python
"""Task 4: Plot a histogram to visualize the distribution of Order_Amount."""

# Sample dataset 4 - Orders Data
data_orders = {
    "Order_ID": [1001, 1002, 1003, 1004, 1005],
    "Customer_ID": [101, 102, 103, 104, 105],
    "Order_Amount": [150, 200, 175, 220, 300],
    "Order_Date": [
        "2023-07-01",
        "2023-07-02",
        "2023-07-03",
        "2023-07-04",
        "2023-07-05",
    ],
}
df_orders = pd.DataFrame(data_orders)
print(df_orders)
```

```python
df_orders['Order_Amount'].plot(kind='hist', bins=4, edgecolor='black')
plt.xlabel('Order_Amount')
plt.title('The distribution of Order Amount')
plt.show()
```

```python
"""Task 5: Plot a pie chart to show the percentage distribution of Department in the employee data."""

# Sample dataset 2 - Employee Data
data_employee = {
    "Employee_ID": [1, 2, 3, 4, 5],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department": ["HR", "IT", "HR", "Sales", "IT"],
    "Salary": [50000, 60000, 55000, 70000, 62000],
}

df_employee = pd.DataFrame(data_employee)
print(df_employee)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

department_counts = df_employee['Department'].value_counts()


plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'lightgreen'])
plt.title('Department Distribution in Employee Data')
plt.axis('equal')  
plt.tight_layout()


plt.show()

```
